```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
z_value = list(map(int, np.random.triangular(1800, 3900, 4500, 1000)))
quantity = list(map(int, np.random.triangular(5, 21, 27, 1000)))

data_dict = {"quantity":quantity,"z_value":z_value}
data = pd.DataFrame(data_dict)
data
```

```python
sns.histplot(data=data, x="quantity", bins=19, kde=True)
plt.show()

sns.histplot(data=data, x="z_value", bins=19, kde=True)
plt.show()
```
